```python
import pandas as pd
```

```python
# Load dataset
df = pd.read_csv('FIFA17_official_data.csv')

df.head()
```

```python
# Display basic information about the dataset
df.info()

# Display DataFrame index
df.index

# Display column names
df.columns
```

```python
#REINDEXING
# Set "Name" as the index
df_reindexed = df.set_index("Name")

# Reset index back to default
df_reindexed = df_reindexed.reset_index()

# Display updated DataFrame
df_reindexed.head()
```

```python
#DROP_ENTRY
# Drop a specific player (e.g., drop first row)
df_dropped = df.drop(df.index[0])

# Display updated DataFrame
df_dropped.head()
```

```python
#SELECTING_ENTRIES
# Select all players from "FC Barcelona"
FC_Barcelona_players = df[df["Club"] == "FC Barcelona"]

# Display first few rows
FC_Barcelona_players.head()
```

```python
#DATA_ALIGNMENT

# Create two small DataFrames
df1 = pd.DataFrame({"Rating": [90, 85, 88]}, index=["Messi", "Ronaldo", "Neymar"])
df2 = pd.DataFrame({"Rating": [92, 87]}, index=["Messi", "Suarez"])

# Adding DataFrames
df_aligned = df1 + df2
df_aligned
```

```python
# RANK AND SORT
# Rank players based on Overall Rating
df["Overall_Rank"] = df["Overall"].rank(ascending=False)

# Sort players by their ranking
df_sorted = df.sort_values(by="Overall_Rank")

# Display top 5 players
df_sorted.head()
```

```python
#SUMMARY_STATISTICS

# Get statistics for all numerical columns
df.describe()

# Get statistics for a specific column
df["Age"].describe()
```

```python
#INDEX_HIERARCHY
# Creating a MultiIndex DataFrame using "Club" and "Nationality"
df_multi = df.set_index(["Club", "Nationality"])

# Accessing data for a specific club and nationality
df_multi.loc[("FC Barcelona", "Spain")].head()
```
